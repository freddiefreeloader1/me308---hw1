```python
import matplotlib.pyplot as plt

clc = [25, 40, 70, 100, 130, 160, 190, 220, 250]                              # clearance values 
h0 = [14.75,22,32.2,41,42.9,51.2,51.3,46.2,50]                              # h0 values
Tmax = [150.88,109.02,69.82,50.32,41.838,37.09,34.87,33.174,32.23]            # T max values                              

plt.plot(clc,h0,"b-", label = "h0 (micrometers)")
plt.plot(clc,Tmax,"g-", label = "T max value")


plt.legend(loc='upper left')
plt.xlabel('clearance (micrometers)')

```
